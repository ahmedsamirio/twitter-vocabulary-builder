```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('twitter_stream.db')
```

```python
users = pd.read_sql_query("SELECT * FROM users", conn)
users.head()
```

There seems to be alot of duplicates, so let's remove them.

```python
users = users.drop_duplicates()
users.head()
```

Even after removing them, we can still see some duplicates, so let's remove duplicates using the screen_name feature.

```python
users.drop_duplicates(subset=['screen_name'], inplace=True)
print(users.shape)

users.head()
```

Now we have 1992 users, let's look into the distributions of followers, friends and statuses.

```python
pd.plotting.scatter_matrix(users[['followers_count', 'friends_count', 'statuses']], figsize=(12, 8));
```

I have an itch to look into the relationshop between statuses and followers count, and I know that it won't be solid, but let's take a look anyway.

```python
users.plot(kind='scatter', x='statuses', y='followers_count', alpha=0.2);
```

I think that if we remove the followers_count outliers, we might find a linear relationship.

```python
tmp = users[['statuses', 'followers_count']]
tmp = tmp[(tmp['followers_count'] < tmp['followers_count'].quantile(0.95))
        & (tmp['followers_count'] > tmp['followers_count'].quantile(0.05))]

tmp.plot(kind='scatter', x='statuses', y='followers_count', alpha=0.2);
```

The relationship isn't strong at all.

```python

plt.hist(tmp['followers_count'], bins=30);
```

```python
plt.hist(tmp['statuses'], bins=30);
```

```python
plt.hist(users['friends_count'], bins=30);
```

Now this plot is definetly weird, because who the hell would have up to 350000 friends (people who you follow) on twitter?

```python
users['friends_count'].plot(kind='box');
```

Let's get to know these strange outliers.

```python
tmp2 = users[users['friends_count'] > users['friends_count'].quantile(0.99)]
tmp2
```

So we can certainly see some twitter accounts that don't represent real persons, and that's why they have these huge number of friends. But also there are some real persons who have a really huge number of friends.
